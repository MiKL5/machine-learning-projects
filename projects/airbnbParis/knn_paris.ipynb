# **Introduction au K plus proches voisins**

#### Afficher le permier élément du dataset

```python
listings = pd.read_csv('paris_airbnb.csv')
print(listings.iloc[0])
```

### **La distance Euclidienne**

<h4>Calculer la distance Euclidienne entre le logoment pouvant accueillir 3 personnes et le premier logement du dataFrame<br>puis, assigner le résultat à `first_distance` avant de l'afficher</h4>

```python
our_acc_value = 3 # personnes dans le logement
first_living_space_value = listings.iloc[0]['accommodates']

first_distance = np.abs(first_living_space_value - our_acc_value)
print(f"La distance pour le premier logement est de {first_distance}.")
```

### **Calculer la distance pour toutes les obesrvations**

* Calculer la distance entre chaque valeur de la colonne '`vccomodates`' de 'listings' et la valreur 3 (le nombre de personnes que peut accueillir le logement) :
    * Utiliser la méthode _appply_ pour calculer la valeur avsolue entre chaue valeur d'`accomodates`  et 3, puis retourner un nouvel objet Series contenant les valeurs de distance.
* Assigner les valeurs de distance à la colonne '`distance`' du dataframe.
* Utiliser la méthode Series `value_counts()`, puis afficher le décompte de chaque valeur unique de la colonne '`distance`'.

```python
new_listings = 3
listings['distance'] = listings['accommodates'].apply(lambda x: np.abs(x - new_listings))
print(listings['distance'].value_counts())
```

### **Randomiser et trier**

* Se placer au même seed `np.random.seed(1)`.
* Randomiser l'ordre des lignes de 'listings' :
    * Retourner un tavleau NumPy de valeurs d'index mélangés par la fonction `np.random.permutation()` ;
    * Retourner un nouveau dataframe contenant l'ordre mélangé via la méthode `loc[]` ;
    * Assigner le nouveau dataframe à 'listings'.
* La randomistation est faite ; trier 'listings' par la colonnes 'distance' grâce à la méthode `sort_values()` ;
* Afficher les dix premières valeurs de la colonne 'price'.

```python
np.random.seed(1)
listings = listings.loc[np.random.permutation(len(listings))]
listings = listings.sort_values('distance')
listings.iloc[0:10]['price']
```

### **Prix moyen**

* Retirer les '`,`' & '`$`' de 'price' :
    * Via l'accesseur `str` pour appliquer la méthode string et la méthode de remplacement des caracgtères à chaque valeur de la colonne précédente : `stripped_commas = listings['price'].str.replace(', ', '')` ;
    * Réitérer pour supprimer les '`$`'.
* Convertir le noiuvel objet en Series contenant les valeurs nettoyées en '`float`' & assigner le résultat à la colonne 'price' de 'listings' ;
* Calculer la moyenne des cinq valeurs de la colonne 'price' et assigner le résultat à la variable 'mean_price'.
Quel est le résultat ?

```python
stripped_commas   = listings['price'].str.replace(',', '')
stripped_dollars  = stripped_commas.str.replace('$', '')
listings['price'] = stripped_dollars.astype(float)
mean_price        = listings.iloc[0:5]['price'].mean()
print(f"Le prix moyen des 5 logements les plus proches est de {mean_price:.2f} dollars.")
```
